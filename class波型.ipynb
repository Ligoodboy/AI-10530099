```python
class AAA10530099:
    """functions of class """
    def __init__(self, x):
        self.x = np.arange(0, 6, 0.1)
        
    def sin (self):
        y1 = np.sin(x)
        plt.plot(self.x, y1)
        plt.title('sin')
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.show()
   
    def cos (self):
        y2 = np.cos(x)
        plt.plot(self.x, y2)
        plt.title('cos')
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.show()
    
    def sin_cos(self):
        y1 = np.sin(x)
        y2 = np.cos(x)
        plt.plot(self.x, y1)
        plt.plot(self.x, y2)
        plt.title('sin & cos')
        plt.xlabel("x") # x軸
        plt.ylabel("y") # y軸
        plt.show()
m = AAA10530099("qqq")
m.sin()
m.cos()
m.sin_cos()
```
